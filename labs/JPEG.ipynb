### DC

```python
import math

def DC_table(fila,columna):
    if fila == 0:
        return 0
    if columna >= 2**(fila-1):
        return columna
    else:
        return -(2**fila)+1+columna
   
   
def DC_table_inv(num):
    if num == 0:
        return (0,0)
    fila = math.floor(math.log2(abs(num)))+1
    if num > 0:
        columna = num
    else:
        columna = num +(2**fila)-1
    return fila,columna
```

```python
def DC_encoder(listaDC):
    lista_diff = [listaDC[0]]
    for i in range(1,len(listaDC)):
        lista_diff.append(listaDC[i] - listaDC[i-1])
    res = ""
    for elem in lista_diff:
        (f,c) = DC_table_inv(elem)
        bin_f = "1"*f+"0"
        bin_c = ""
        if f != 0: bin_c = '{0:0b}'.format(c).zfill(f)
        res += bin_f + bin_c
    return res

def DC_decoder(bin_DC):
    res = []
    i = 0
    while i < len(bin_DC):
        f = 0
        c = 0
        bit = bin_DC[i]
        while bit == "1":
            f+=1
            i+=1
            bit = bin_DC[i]
        i+=1#skip the zero
        if f == 0:
            c = 0
        else:
            c = int(bin_DC[i:i+f],2)
        i+=f
        diff = DC_table(f,c)
        num = diff
        if res != []: num += res[-1]
        res.append(num)
    return res
        
```
