**Definir una función** `IntegerArithmeticCode(mensaje, alfabeto, frecuencias)` **que dado un $mensaje$ en el que se ha utilizado el $alfabeto$ con las correspondientes $frecuencias$ devuelva el resultado de codificar dicho mensaje usando codificación aritmética entera en el intervalo de trabajo $[0,R)$.**


Frecuencias: $f(1)$,..., $f(n)$, $f(i)$ entero

Frecuencias acumuladas:

* $F(0)=0$, 
* $F(i)=\sum_{k=1}^{i}f(k)$,
* $T=F(n)$ suma total de frecuencias


Intervalo de trabajo: $[0,R)$, $R=2^k$, $R>4T$


```python
import math
def IntegerArithmeticCode(mensaje, alfabeto, frecuencias):
    T = sum(frecuencias)
    power = math.ceil(math.log(4*T,2))
    R = 1<<power
    print(R)
    l = 0
    u = R-1
    
    d = {alfabeto[i]:i+1 for i in range(len(alfabeto))}
    s = 0
    rango = [s]
    for elem in frecuencias:
        s += elem
        rango += [s]
        
    mfin = ""
    S3 = 0
    for elem in mensaje:
        x = d[elem]
        lnew = l+math.floor(((u-l+1)*rango[x-1])/T)
        unew = l+math.floor(((u-l+1)*rango[x])/T) - 1
        l = lnew
        u = unew
        E3 =  (l>>(power-2)) == 1 and (u>>(power-2)) == 2 
    
        while((l>>(power-1)) == (u>>(power-1)) or E3):
            b = l>>(power-1)
            
            if((l>>(power-1)) == (u>>(power-1))):
                mfin += str(b)
                l = (l<<1) & (R-1)
                u = (((u<<1) & (R-1)) | 1)
                
                if S3 > 0:
                    mfin += str(1 - b)*S3
                    S3 = 0
            if(E3):
                l = (l<<1) & (2**power-1) ^ (R>>1)
                u = (((u<<1) & (2**power-1)) | 1) ^ (R>>1)
                S3 += 1
                
            E3 =  (l>>(power-2)) == 1 and (u>>(power-2)) == 2  
            
    if(l>>(power-2) == 0):
            mfin += "01"
    else:
            mfin += "10"
    
    mfin += format(l, "#0"+str(power+2)+"b")[2:]
       
            
    print(mfin)
    return mfin       
 
```

**Definir una función** `IntegerArithmeticDecode(codigo, longitud_mensaje, alfabeto, frecuencias)` **que dado un mensaje codificado $codigo$ usando codificación aritmética entera en el intervalo de trabajo $[0,R)$  en el que se ha utilizado el $alfabeto$ con las correspondientes $frecuencias$ devuelva el mensaje original de de longitud $longitud\_mensaje$.**



```python
   
def IntegerArithmeticDecode(codigo, longitud_mensaje, alfabeto, frecuencias):
    T = sum(frecuencias)
    power = math.ceil(math.log(4*T,2))
    R = 1<<power
    print(R)
    l = 0
    u = R-1
    i = 0
    tag = codigo[i:i+power]
    i += power
    d = {alfabeto[i]:i+1 for i in range(len(alfabeto))}
    s = 0
    rango = [s]
    for elem in frecuencias:
        s += elem
        rango += [s]
    print(rango)
    mensaje = ""
  
    while len(mensaje) < longitud_mensaje:

        k = 0
        val = math.floor(((int(tag,2)-l+1)*T - 1)/(u-l+1))
        print(val)
        while  val >= rango[k]:
            k = k+1
            
        mensaje += alfabeto[k-1]
        lnew = l+math.floor(((u-l+1)*rango[k-1])/T)
        unew = l+math.floor(((u-l+1)*rango[k])/T) - 1
        l = lnew
        u = unew     
                      
        while((l>>(power-1)) == (u>>(power-1)) or ((l>>(power-2)) == 1 and (u>>(power-2)) == 2) ):
            print("L ", bin(l))
            print("U ", bin(u))  
            if((l>>(power-1)) == (u>>(power-1))):
                print("Change1", mensaje)            
                l = (l<<1) & (R-1)
                u = (((u<<1) & (R-1)) | 1)
                print("L ", bin(l))
                print("U ", bin(u))  
                tag = tag[1:]+codigo[i:i+1]
                i += 1 
            if((l>>(power-2)) == 1 and (u>>(power-2)) == 2):
                print("Change2", mensaje)  
                l = ((l<<1) & (R-1)) ^ (R>>1)
                u = (((u<<1) & (R-1)) | 1) ^ (R>>1)
                print("L ", bin(l))
                print("U ", bin(u))  
                tag = tag[1:]+codigo[i:i+1]
                tag = str(1-int(tag[0])) + tag[1:]
                i += 1

    return mensaje[:longitud_mensaje]
```
